```python
#ライブラリの読み込み
import numpy as np
import numpy.random as random
import matplotlib as mpl
#Seabornは綺麗に図示できる
import seaborn as sns

#pyolotにはpltの別名で実行できるようにする
import matplotlib.pyplot as plt

#jupyter上でグラフを表示させるために必要なマジックコマンド
%matplotlib inline
```

散布図

```python
#seed値の固定
random.seed(0)

#x軸のデータ
x = np.random.randn(30)

#y軸のデータ
y = np.sin(x) + np.random.randn(30)

#グラフの大きさ指定
plt.figure(figsize=(20,6))

#グラフの描写
plt.plot(x,y,'o')

#以下でも散布図が書ける
#plt.scatter(x,y)

#タイトル
plt.title('Title Name')
#xの座標名
plt.xlabel('x')
#yの座標名
plt.ylabel('Y')

#grid(グラフの中にある縦線と横線)の表示
plt.grid(True)
```

連続曲線

```python
#seed値の固定
random.seed(0)

#データの範囲
numpy_data_x = np.arange(1000)

#乱数の発生と積み上げ
numpy_random_data_y = np.random.randn(1000).cumsum()

#グラフの大きさを指定
plt.figure(figsize=(20,6))

#labelとlegendでラベルを付けることが可能
plt.plot(numpy_data_x,numpy_random_data_y, label = 'Label')
plt.legend()

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

グラフの分割

```python
#subplotを使うとグラフを複数に分けることができる。今回は縦に２つ(２行１列)

#グラフの大きさを指定
plt.figure(figsize=(20,6))

#2行1列のグラフの１つ目(y=sinx)
#linspace(-10,10,100)は-10から10までの数を100個に分割した数字リスト
plt.subplot(2,1,1)
x = np.linspace(-10,10,100)
plt.plot(x,np.sin(x))

#2行1列のグラフの２つ目(y=2sinx)
plt.subplot(2,1,2)
y = np.linspace(-10,10,100)
plt.plot(y,np.sin(2*y))

plt.grid(True)
```

関数グラフの描画

```python
#関数の定義(f(x)=x^2+2x+1)
def my_function(x):
    return (x**2 + 2*x +1)

x = np.arange(-10,10)
plt.figure(figsize=(20,6))
plt.plot(x,my_function(x))
plt.grid(True)
```

↑解の位置が大体わかる

```python
#f(x)の最小値を求める(Brent法):Scipyの基礎より
from scipy.optimize import minimize_scalar
print(minimize_scalar(my_function, method = 'Brent'))
```

ヒストグラム

```python
#seed値の固定
random.seed(0)

#グラフの大きさを指定
plt.figure(figsize=(20,6))

#ヒストグラムの描写
#対象となるデータ,ビンの数(幅、個数),範囲
#randn:標準正規分布。平均０、標準偏差１の乱数
plt.hist(np.random.randn(10 ** 5) * 10 + 50, bins=60, range=(20,80))

plt.grid(True)
```
